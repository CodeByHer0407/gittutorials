# Flask Framework — Introduction
---

## What is Flask?

Flask is a **web framework written in Python** that lets you build web applications and APIs.

It is built on two core components:
- **WSGI** — Web Server Gateway Interface (protocol between web server and web app)
- **Jinja2** — Template engine for creating dynamic web pages

---

## How a Flask App Works — The Big Picture

```
Client Request
     │
     ▼
  Web Server  (AWS EC2 / Azure / Apache / IIS)
     │
     │  WSGI Protocol
     ▼
  Flask App   (/home, /api, /dashboard ...)
     │
     ▼
  Response sent back to client
```

**Real world analogy:**
Think of a restaurant.
- The **client** is the customer placing an order
- The **web server** is the waiter taking the order
- **WSGI** is the communication protocol between waiter and kitchen
- The **Flask app** is the kitchen that prepares and returns the response

---

## WSGI — Web Server Gateway Interface

- A **protocol/standard** that defines how a web server communicates with a Python web app
- Without WSGI, the web server wouldn't know how to talk to Flask
- Flask implements WSGI under the hood — you don't write WSGI code directly

**Real world example:**
When you visit `www.somesite.com/home`, the web server (e.g. Apache) receives the request and uses WSGI to forward it to your Flask app, which handles the `/home` route and sends back a response.

---

## Jinja2 — Template Engine

- Jinja2 lets you create **dynamic web pages** by injecting data into HTML templates
- Data can come from any source: SQL DB, CSV, ML Model, MongoDB, etc.

```
Data Source  ──►  Web Template  ──►  Web Page
(SQL / ML Model / MongoDB)
```

**Real world example:**
An e-commerce site shows each user their own order history. The HTML template is the same for everyone, but Jinja2 fills in each user's specific data dynamically from the database.

```html
<!-- Jinja2 template example -->
<h1>Welcome, {{ username }}!</h1>
<p>Your order total: {{ order_total }}</p>
```

## Flask vs FastAPI — Quick Note

| | Flask | FastAPI |
|---|---|---|
| Type | Synchronous (WSGI) | Asynchronous (ASGI) |
| Use case | Web apps, simple APIs | High-performance APIs, AI backends |
| Auto docs | No | Yes (Swagger UI built-in) |
| Industry use | General web dev | ML/AI backends, modern APIs |

> Flask is a great stepping stone — understanding routes, HTTP, and templates here makes FastAPI much easier to pick up.

---

## Key Takeaways

- Flask is a Python web framework built on **WSGI + Jinja2**
- WSGI = protocol that connects web server to Flask app
- Jinja2 = template engine that creates dynamic pages from data sources
- Flask handles routes like `/home`, `/api`, `/dashboard` and returns responses
- Relevant for AI backends: Flask can serve ML models as APIs (predict endpoints)
